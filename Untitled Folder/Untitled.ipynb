```python
def uniq_columns(data):
    columns = data.columns
    add_columns = []
    for c in data.columns:
        uniq_len = len(list(data[c].unique()))
        if(uniq_len > 20):
            add_columns.append(c)
    return add_columns 


cols = uniq_columns(df)
#df[[uniq_columns(df)]]
cols
```

```python
cont_df = df.select_dtypes('number')
cols = uniq_columns(cont_df)
uniq_cols = cont_df[[col for col in cols]]
uniq_cols
```

availability_365 = off the market or no data
availabililty_365 and number of reviews = 0 
after removing we can then bin the availability

room_type = 3 different categories
minimum nights = 1-3, 4-14,15-30, 30+, 90+

```python
rm_dummies = pd.get_dummies(df['room_type'], prefix = 'rm_type', drop_first=True)
```

```python
data = pd.concat([rm_dummies, uniq_cols, ], axis = 1 )
```

```python
data.rename(columns = {'rm_type_Private room': 'private_room', "rm_type_Shared room": 'shared_room'}, inplace = True)
data
```

```python
data.info()
```

```python
data['reviews_per_month'].fillna(data['reviews_per_month'].mean(), inplace = True)
data.drop(columns= ['id', 'host_id'],axis = 1, inplace = True)
```

```python
from scipy import stats
data = data[(np.abs(stats.zscore(data)) < 3).all(axis=1)]
```

```python
stats.zscore(data['price'])
```

```python
formula = 'price ~ private_room+shared_room+number_of_reviews+availability_365+latitude+longitude+reviews_per_month+calculated_host_listings_count'
model = ols(formula = formula, data = data).fit()
model.cross_val
```

```python
import sklearn as sk


log_df = pd.DataFrame([])
log_df['number_of_reviews_log'] = np.log(1 + data['number_of_reviews'])
log_df['reviews_per_month_log'] = np.log(1 + data['reviews_per_month'])
log_df['availability_log'] = np.log(1 + data['availability_365'])
log_df['price'] = data['price']
log_df['shared_room'] = data['shared_room']
log_df['private_room'] = data['private_room']
log_df['latitude'] = data['latitude']
log_df['longitude'] = data['longitude']



formula = 'price ~ number_of_reviews_log+reviews_per_month_log+availability_log+shared_room+latitude+longitude'
model = ols(formula = formula, data = log_df).fit()
model.summary()
```
